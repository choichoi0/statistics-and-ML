### Showing multiple graphs at once

```python
x1 = np.linspace(-5, 0)
x2 = np.linspace(0, 5)

y1 = np.cos(x1)
y2 = np.cos(x2)
```

```python
# Show multiple graphs at once
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r--')
plt.title('First figure')

plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'b-')
plt.title('Second figure')
```

```python
# Sharing axis
ay1 = plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r--')
plt.title('First figure')

plt.subplot(1, 2, 2, sharey=ay1)
plt.plot(x2, y2, 'b-')
plt.title('Second figure')
plt.yticks(visible=False)
plt.tick_params(length=0)
```
