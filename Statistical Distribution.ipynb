## Normal distribution

### pdf : probability density function

```python
from scipy import stats

rv = stats.norm(0, 1)# Mean 0, standard deviation 1
x = np.linspace(-5, 5, 100) # Create 100 pieces from -5 to 5 at equal intervals
y1 = rv.pdf(x)
plt.plot(x, y1)
plt.show()
```

```python
rv = stats.norm(0, 2) # Mean 0, standard deviation 2
x = np.linspace(-5, 5, 100) 
y2 = rv.pdf(x)
plt.plot(x,y2)
plt.show()
```

#### Practice 1
A company surveyed the working period of employees, and it is said that the working period of employees follows a normal distribution with an average of 11 years and a standard deviation of 4 years.  
Find the percentage of employees who have worked for this company for more than 14 years.

```python
rv = stats.norm(11, 4)
prob = 1 - rv.cdf(14)
print(f"{prob * 100}%")
```

#### Practice 2. 
It is said that the life of a bulb company's bulb follows a normal distribution with an average of 800 days and a standard deviation of 30 days.   
Find the probability that the bulb has a lifespan of less than 760 days.

```python
rv = stats.norm(800, 30)
prob = rv.cdf(760)
print(f"{prob * 100}%")
```

#### Practice 3
It is said that the math grades of a high school senior follow a normal distribution with an average of 70 points and a variance of 64 points.
Find the probability that your math score is 80 or higher and 90 or lower.

```python
rv = stats.norm(70, math.sqrt(64))
prob = rv.cdf(90) - rv.cdf(80)
print(f"{prob * 100}%")
```

## T distribution

```python
# Drawing a T Distribution Graph
rv = stats.t(1) # degree of freedom 1
x = np.linspace(-5, 5, 100)
y = rv.pdf(x)
plt.plot(x, y)
plt.show()
```

```python
# Comparison of graphs according to degrees of freedom
x = np.linspace(-5, 5, 100) 

for i in [1, 2, 3, 999]:
    rv = stats.t(i)
    y = rv.pdf(x)
    plt.plot(x, y)

rv = stats.norm(0, 1)
y = rv.pdf(x)
plt.plot(x, y)
plt.legend(["1", "2", "3", "999", "normal"])
plt.show()
```

## Central limit theorem

```python
# Throwing two dice n times
def two_dice(n):
    result_list = [0] * 13 # List to store the sum of two dice
    for i in range(n):
        dice1 = random.randint(1, 6)
        dice2 = random.randint(1, 6)
        res = dice1 + dice2
        result_list[res] += 1 
        
    for i in range(13):
        result_list[i] /= n
        
    return result_list

# plotting)
def plot_result(result_list):
    plt.bar(range(2, 13), result_list[2:], tick_label=range(2, 13)) # bar graph 
    plt.ylim(0, 0.3) # y axis: 0 ~ 0.3
    plt.ylabel('prob') # y axis name: prob
    plt.xlabel('2 dice sum') # x axis name: 2 dice sum
    plt.show()
```

```python
# Throw two dice 1000000 times using the function defined above

n = 1000000
result_list  = two_dice(n)
plot_result(result_list)
```

## F distribution

```python
rv = stats.f(1, 5) # degree of freedom 1, 5
x = np.linspace(0, 5, 100)
y = rv.pdf(x)
plt.plot(x, y)
plt.xlim(left = 0)
plt.ylim(bottom = 0)
plt.show()
```

```python
# Comparison of F distribution graphs according to degrees of freedom
rv = stats.f(1, 1) # degree of freedom 1, 1
x = np.linspace(0, 5, 100) 

for i in [1, 5, 10]:
    rv = stats.f(i, i)
    y = rv.pdf(x)
    plt.plot(x, y)

y = rv.pdf(x)
plt.plot(x, y)
plt.xlim(left = 0)
plt.ylim(bottom = 0)
plt.legend(["1, 1", "5, 5", "10, 10"])
plt.show()
```

## Practice

### Normal Distribution
- The average of statistical test results over the last five years was 70 points, and the standard deviation was 10 points.
- If only the top 10% get an A grade, how many test scores do I have to get to get an A grade?

```python
#   
rv = stats.norm(70, 10)

min = -1

#  10%   
for i in range (50, 101):
    if rv.cdf(i) > 0.9:
        min = i
        break

# , cdf (ppf)   .
min_by_ppf = rv.ppf(0.9)

print(f"A    : {min}")
print(f"A    ( ): {math.ceil(min_by_ppf)}")
```

### T distribution
- The average amount of milk for 10 ranches was investigated to determine the average amount of milk produced by cows at dairy ranches in Gyeonggi Province.
- As a result, the mean was 35 kg and the standard deviation was 5 kg.
- At this time, in order to be in the top 5% in Gyeonggi-do, the average milk content should be at least how many kilograms?

```python
#    ( 9)
rv = stats.t(9)

#  5%   
#     .
std_min = rv.ppf(0.95)

#      X ,
min = std_min * (5 / math.sqrt(10)) + 35

print(f"   5%   : {min}kg")
```

- What percentage of the total farms that average over 40 kg?

```python
#    ( 9)
rv = stats.t(9)

#  t 
std_t = (40 - 35) / (5 / math.sqrt(10))

# cdf  
percentage = (1 - rv.cdf(std_t)) * 100

print(f"Farms averaging more than 40 kg represent {percentage}% of the total.")
```
