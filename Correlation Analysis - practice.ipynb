# Correlation Analysis - practice

- The following is data related to car accidents provided by the seaborn library.
- Find the correlation between speeding (*'speed'*) and accident (*'total'*) through the correct correlation analysis process.
    - Draw a scatterplot with labels.
    - Calculate the covariance, correlation coefficient.
    - Perform a correlation coefficient test.

```python
car = sns.load_dataset('car_crashes')
display(car)
```

```python
X = car.speeding.values
Y = car.total.values
plt.scatter(X, Y)
plt.xlabel('Speeding')
plt.ylabel('Total')
```

```python
np.cov([X, Y])[0][1]
```

```python
np.corrcoef([X, Y])[0][1]
```

```python
from scipy import stats
stats.pearsonr(X, Y)
```
