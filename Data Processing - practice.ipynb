# Data Processing Practice

- The following are statistics on deaths from cancer by state in the United States.  
https://drive.google.com/file/d/1SkCbZdnzng1a43m_vRRrVbmeFOPcyIU0/view?usp=sharing
- Download the above statistics under the name 'cancer.csv' using the gdown package and store them in the variable 'cancer.'
- Pre-process according to the following requirements.
  - Use the state name of the United States as an index.
  - Fill in the missing values with the average value of the data.
  - Treat outliers in an IQR manner.  
  - Use variable numbers to process categorical data.

```python
gdown.download("https://drive.google.com/uc?id=1SkCbZdnzng1a43m_vRRrVbmeFOPcyIU0", "cancer.csv", quiet=False)
cancer = pd.read_csv('cancer.csv')
```

```python
cancer.head()
```

```python
cancer = cancer.set_index('State')
cancer.head()
```

```python
cancer.isnull().any()
```

```python
cancer=cancer.fillna(cancer.mean())
display(cancer.fillna(cancer.mean()))
```

```python
def iqr_outlier(data):
    for i in range(len(data.iloc[1])):
        if type(data.iloc[0, i]) != np.int64 and type(data.iloc[0, i]) != np.float64:
            continue
        q1, q3 = np.percentile(data.iloc[:, i], [25, 75])
        print(f"q1: {q1}")
        print(f"q3: {q3}")

        iqr = q3 - q1
        lower = q1 - (iqr * 1.5)
        upper = q3 + (iqr * 1.5)
        print(f"lower: {lower}")
        print(f"upper: {upper}")

        data = data[data.iloc[:, i] < upper]
        data = data[data.iloc[:, i] > lower]
    return data

wo_outlier = iqr_outlier(cancer)
wo_outlier.plot(legend=False, xlabel="X", ylabel="Y")
```

```python
cancer = pd.get_dummies(cancer, drop_first=True, prefix=['range'])
cancer.head(n=10)
```
